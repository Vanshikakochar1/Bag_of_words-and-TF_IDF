#Bag of Words & TF-IDF

# Import Dependencies

```python
import pandas as pd
import nltk
import re
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

# Read Data

```python
df = pd.read_csv("tweets.csv")
df.head()
```

```python
tweets = df["Text"]
```

```python
tweets.head()
```

# Preprocess

```python
def preprocess(text):
    text = text.lower()
    text = re.sub(r"\W", " ", text)
    text = re.sub(r"\s+", " ", text)
    return text
```

```python
tweets = tweets.apply(preprocess)
tweets.head()
```

# Bag of Words

```python
tweets = df["Text"]
```

```python
def preprocess(text):
    text = text.lower()
    text = re.sub(r"\W", " ", text)
    text = re.sub(r"\s+", " ", text)
    return text
```

```python
tweets = tweets.apply(preprocess)
tweets.head()
```

```python
vectorizer = CountVectorizer()
```

```python
vectorizer.fit(tweets)
```

```python
vectorizer.get_feature_names_out()
```

```python
vectorizer.vocabulary_
```

```python
vector = vectorizer.transform(tweets)
```

```python
print("Encoded Document is:")
print(vector.toarray())
```

```python
print([vector.toarray()[0]])
print(vectorizer.inverse_transform([vector.toarray()[0]]))
```

# TF-IDF

```python
tf_idf_vectorizer = TfidfVectorizer()
```

```python
tf_idf_vectorizer.fit(tweets)
```

```python
tf_idf_vectorizer.get_feature_names_out()
```

```python
tf_idf_vectorizer.vocabulary_
```

```python
tf_idf_vector = tf_idf_vectorizer.transform(tweets)
```

```python
print("Encoded Document is:")
print(tf_idf_vector.toarray())
```

```python
print([tf_idf_vector.toarray()[0]])
print(tf_idf_vectorizer.inverse_transform([tf_idf_vector.toarray()[0]]))
```
